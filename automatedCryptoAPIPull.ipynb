```python
import pandas as pd

dfFromCSV = pd.read_csv(r"/Users/thomasmyrseth1/Documents/fritid/koding/webscrape/coinMarkedCapWebScrape/coinMarkedCapWebScrape.csv")
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

dfFromCSV
```

```python
dfFromCSV = dfFromCSV.groupby("name", sort=False)[["quote.USD.percent_change_1h", "quote.USD.percent_change_24h", "quote.USD.percent_change_7d", "quote.USD.percent_change_30d", "quote.USD.percent_change_60d", "quote.USD.percent_change_90d"]].mean() #average value of all percentiles

#velger ut hvilken data jeg vil se ^
```

```python
df4=dfFromCSV.stack()  #gjør om til kolonner

df5 = df4.to_frame(name = "percentChange")

df5
```

```python
index = pd.Index(range(df5.shape[0])) #90 kolonner, ikke i bruk

df6 = df5.reset_index()

df6 = df6.rename(columns={"name": "currency", "level_1": "time_interval", "percentChange": "percent_change"})

df6
```

```python
#endrer kollonenavnene under time_interval

df6["time_interval"] = df6["time_interval"].replace(["quote.USD.percent_change_1h", "quote.USD.percent_change_24h", "quote.USD.percent_change_7d", "quote.USD.percent_change_30d", "quote.USD.percent_change_60d", "quote.USD.percent_change_90d"], ["1h", "24h", "7d", "30d", "60d", "90d"])

df6
```
